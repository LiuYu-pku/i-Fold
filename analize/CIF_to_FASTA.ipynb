```python
from Bio.PDB import PDBParser, MMCIFParser, Superimposer  
from Bio.PDB.Polypeptide import three_to_index, index_to_one
from Bio.PDB import PDBIO
from Bio import PDB
import numpy as np  
import os
import re
import pandas as pd
```

```python
filePath = '/home/liuyu/codes/ifold/my_works/testset/mmcif_test_250409'
cif_lst = os.listdir(filePath)
cif_lst
```

```python
import warnings
warnings.filterwarnings("ignore")
for item in cif_lst:
    label_cif = f'/home/liuyu/codes/ifold/my_works/testset/mmcif_test_250409/{item}'
    parser_cif = MMCIFParser()
    structure1 = parser_cif.get_structure("protein1", label_cif)
    model1 = structure1[0]
    chain1 = [chain for chain in model1][0]
    # 保存单链PDB
    chain_id = chain1.get_id()
    prot_id = item.split('.')[0]
    io = PDBIO()
    io.set_structure(chain1)
    try:
        io.save(f'./SingleChainPdb_250409/{prot_id}_{chain_id}.pdb')
    except:
        pass
    # 生成fasta
    try:
        aa_lst = []
        for residue in chain1:
            if PDB.is_aa(residue):
                aa_lst.append(index_to_one(three_to_index(residue.resname)))
        seq_str = ''.join(aa_lst)
        with open(f'./fasta_test_250409/{prot_id}_{chain_id}.fasta', 'w') as f:
            f.write(f'>{prot_id}_{chain_id}\n')
            f.write(seq_str)
    except:
        pass
```
