```python
import os
```

```python
filePath = '/home2/liuyu/ifold_trainset/alignment_data/alignments/'
align_file_lst = os.listdir(filePath)
len(align_file_lst)
```

```python
seq_list = []
with open('/home2/liuyu/ifold_trainset/alignment_data/all-seqs.fasta') as f:
    for line in f:
        if line.startswith('>'):
            line = line.strip().strip('>')
            seq_list.append(line)
```

```python
len(seq_list)
```

```python
intersect_lst = list(set(align_file_lst) & set(seq_list))
len(intersect_lst)
```

```python
from tqdm import tqdm
```

```python
perfect_align_lst = []
for item in tqdm(align_file_lst):
    fp = f'/home2/liuyu/ifold_trainset/alignment_data/alignments/{item}'
    file_count = len(os.listdir(fp))
    if file_count == 4:
        perfect_align_lst.append(item)
```

```python
len(perfect_align_lst)
```
