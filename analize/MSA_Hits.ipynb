```python
import os
import pandas as pd
import numpy as np
```

```python
def count_sequences_in_a3m(file_path):
    """统计a3m文件中的'>'数量"""
    try:
        with open(file_path, 'r') as f:
            return f.read().count('>')
    except Exception as e:
        return np.nan
```

```python
root_dir = '/home/liuyu/codes/ifold/my_works/testset/alignment_test'

id_lst = []
count_lst = []

# 遍历根目录下所有子目录
for protein_id in os.listdir(root_dir):
    protein_dir = os.path.join(root_dir, protein_id)
    
    # 确保是目录
    if not os.path.isdir(protein_dir):
        continue
        
    a3m_path = os.path.join(protein_dir, 'bfd_uniclust_hits.a3m')
    
    if os.path.isfile(a3m_path):
        count = count_sequences_in_a3m(a3m_path)
    else:
        count = np.nan
    print(count)
    id_lst.append(protein_id)
    count_lst.append(count)
```

```python
df = pd.DataFrame({'prot_id':id_lst, 'MSA_count':count_lst})
df
```

```python
df_sorted = df.sort_values(by='MSA_count')
df_sorted
```

```python
df_sorted.to_csv('msa_count.csv', index=False)
```
